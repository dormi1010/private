```python
import urllib.request
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/maskot1977/ipython_notebook/master/toydata/shop_distance.txt'
```

```python
urllib.request.urlretrieve(url, 'ice_cream.txt')
col1 = []
col2 = []
col3 = []
for i, line in enumerate(open('ice_cream.txt', encoding='utf-8')):
    if i == 0:
        continue
    c = line.split()
    col1.append(int(c[0])) 
    col2.append(int(c[1]))
    col3.append(int(c[2]))
```

```python
plt.scatter(col2, col3, alpha=0.5)
plt.title("Numbers of customers in 12shops")
plt.xlabel("Distance from station (m)")
plt.ylabel("Customers")
for shop, x, y in zip(col1, col2, col3):
    plt.text(x, y, shop, alpha=0.8, size=12)
plt.grid()
plt.show()
```
