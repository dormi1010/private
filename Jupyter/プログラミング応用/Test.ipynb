```python
import copy
def dijkstra(adj,s):
    vs = adj.keys()
    dist = {v:float('inf') for v in vs}
    root = {v:[s] for v in vs}
    dist[s] = 0 # s 0
    fixed = set() # 
    while True:
        min_dist = float('inf')
        for v in vs:
            if (v not in fixed) and min_dist>dist[v]:
                min_dist=dist[v]
                min_v = v
        if min_dist==float('inf'): break
        fixed.add(min_v)
        for (u,d) in adj[min_v]:
            if dist[u] > dist[min_v]+d:
                root[u] = copy.deepcopy(root[min_v])
                root[u].append(u) #
                dist[u] = dist[min_v]+d # 
            
    print(dist)
```
