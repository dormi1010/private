```python
corr
```

```python
# 課題7.6
corr_dict = corr.to_dict()
corr2 = []
for x in corr_dict.keys():
    for y in corr_dict[x].keys():
        if x >= y:
            continue
        corr2.append([x, y, corr_dict[x][y]])
```

```python
df2 = pd.DataFrame(corr2)
```

```python
df2.sort_values(2, ascending=False)
```

```python
df2[df2[2] >= 0.4]
```

```python
df2[df2[2] <= (-0.2)]
```

```python
'''
「ナッツ系」や「チョコレート系」、「和風」といった、類似の性質をもつ味同士では正の相関が強い。
そのため、これらのサンプルはそれぞれひとまとまりのグループとして考えられる。
一方で、バニラは「ミント類」や「ナッツ類」と負の相関がみられるので、これらの味は好みの層が比較的分離していると考えられる。
'''
```
